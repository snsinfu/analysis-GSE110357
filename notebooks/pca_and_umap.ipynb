```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from sklearn.decomposition import PCA
from umap import UMAP
```

```python
plt.style.use("./note.mplstyle")
```

```python
scRNA_data = pd.read_csv("../data/processed/scRNAseq-GSE110357.csv", index_col="gene")
```

## PCA

```python
X = np.log(1 + scRNA_data.values.T)
pca = PCA(n_components=2).fit(X)
```

```python
gene = "Vim"

fig, ax = plt.subplots(figsize=(2, 2))

x, y = pca.transform(X).T
c = np.log(1 + scRNA_data.loc[gene])

ax.scatter(x, y, c=c, s=5, edgecolor="none")

ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
ax.set_title(gene)

pass
```

## UMAP

```python
X = np.log(1 + scRNA_data.values.T)
umap = UMAP(n_components=2).fit(X)
```

```python
gene = "Vim"

fig, ax = plt.subplots(figsize=(2, 2))

x, y = umap.transform(X).T
c = np.log(1 + scRNA_data.loc[gene])

ax.scatter(x, y, c=c, s=5, edgecolor="none")

ax.set_xlabel("UMAP1")
ax.set_ylabel("UMAP2")
ax.set_title(gene)

pass
```
